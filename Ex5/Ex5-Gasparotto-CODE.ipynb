```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.animation import FuncAnimation, PillowWriter

from harmonic_oscillator import HarmonicOscillator
from debugger import vector_checkpoints

import warnings
warnings.filterwarnings("ignore")
```

```python
# Latex-like font for plots
# Note: if LaTeX is not installed, do not run this cell
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "cmr10",
    "text.latex.preamble": r"\usepackage{amsmath, amssymb}",
    "font.size": 16})
```

```python
# Controls

# Space discretization
a = -8.
b = 8.
N_slices = 1000

# Time discretization
ti = 0.
tf = 20.
Nt_slices = 1000

# Simulation
simulation_time = 20.

if a>=b:
    raise TypeError("ERROR: a should be lower than b!")

if ti>tf:
    raise TypeError("ERROR: ti should be lower than tf!")
```

```python
# Define Harmonic oscillator objects with different final times
# to inspect three different regimes

ho = HarmonicOscillator(a=a, b=b, npts=N_slices,
                        ti=ti, tf=5, npts_t=Nt_slices,
                        simulation_time=simulation_time,
                        time_dependent=True) # final time half than simulation time

h1 = HarmonicOscillator(a=a, b=b, npts=N_slices,
                        ti=ti, tf=tf, npts_t=Nt_slices,
                        simulation_time=simulation_time,
                        time_dependent=True) # final time equal than simulation time

h2 = HarmonicOscillator(a=a, b=b, npts=N_slices,
                        ti=ti, tf=50, npts_t=Nt_slices,
                        simulation_time=simulation_time,
                        time_dependent=True) # final time double than simultation time
```

```python
# Extract eigenvectors for ho
eigvals, eigvecs = ho.eigenproblem(k=10, t=0.)
psi0 = eigvecs[:,0]
psi1 = eigvecs[:,1]
psi2 = eigvecs[:,2]

# Check for normalizations
vector_checkpoints(0, psi0, ho.dx)
vector_checkpoints(0, psi1, ho.dx)
vector_checkpoints(0, psi2, ho.dx)

# Compute time evolution
psi_t_ho, T_grid_ho = ho.time_evo(psi0, simulation_time=simulation_time)
psi_t_ho = np.array(psi_t_ho)
psi_t_ho.shape
```

```python
# Extract ground state eigenvector for h1
eigvals, eigvecs = h1.eigenproblem(k=1, t=0.)
psi0_h1 = eigvecs[:,0]

# Check for normalization
vector_checkpoints(0, psi0_h1, h1.dx)

# Compute time evolution
psi_t_h1, T_grid_h1 = h1.time_evo(psi0_h1, simulation_time=simulation_time)
psi_t_h1 = np.array(psi_t_h1)
psi_t_h1.shape
```

```python
# Extract ground state eigenvector for h2
eigvals, eigvecs = h2.eigenproblem(k=1, t=0.)
psi0_h2 = eigvecs[:,0]

# Check for normalization
vector_checkpoints(0, psi0_h2, h2.dx)

# Compute time evolution
psi_t_h2, T_grid_h2 = h2.time_evo(psi0_h2, simulation_time=simulation_time)
psi_t_h2 = np.array(psi_t_h2)
psi_t_h2.shape
```

```python
# Compute period and create the grid
T_osc = (2 * np.pi) / ho.omega
times_sec = [T_osc/4, T_osc/2, (3/4)*T_osc, T_osc, (5/4)*T_osc, (3/2)*T_osc, 2*T_osc, 4*T_osc, (10/4)*T_osc]

# Time evolution plot for psi=0
fig = plt.figure(figsize=(13, 7)) 
gs = fig.add_gridspec(2, 2, height_ratios=[1, 1], width_ratios=[1, 1], hspace=0.3)

fig.suptitle(f"$T$={ho.simulation_time}, $t_f={ho.tf}$")

# pdf
ax1 = fig.add_subplot(gs[0, :])
ho.plot_time_evo(ax1, psi_t_ho, times_sec, transform_func=lambda psi: np.abs(psi)**2, y_label=r"$|\psi(x)|^2$", plot_legend=True)

# Real part
ax2 = fig.add_subplot(gs[1, 0])
ho.plot_time_evo(ax2, psi_t_ho, times_sec, transform_func=np.real, y_label=r"$Re(\psi(x))$")

# Imaginary part
ax3 = fig.add_subplot(gs[1, 1])
ho.plot_time_evo(ax3, psi_t_ho, times_sec, transform_func=np.imag, y_label=r"$Im(\psi(x))$")

fig.tight_layout()
plt.savefig("time_evo_plot_ho.pdf")
plt.show()
```

```python
# Simulation parameters to loop over
simulations = [
    (ho, psi_t_ho, T_grid_ho, exp_value_ho, r"Fast ($t_f=5$)"),
    (h1, psi_t_h1, T_grid_h1, exp_value_h1, r"Intermediate ($t_f=20$)"),
    (h2, psi_t_h2, T_grid_h2, exp_value_h2, r"Adiabatic ($t_f=50$)")
]

# Calculation of global maximum for a common color scale
global_max_psi2 = max(
    np.max(np.abs(psi_t_ho)**2),
    np.max(np.abs(psi_t_h1)**2),
    np.max(np.abs(psi_t_h2)**2)
)

# Plot
fig, axes = plt.subplots(3, 1, figsize=(10, 12), sharex=True, sharey=True)
plt.subplots_adjust(hspace=0.1)
im = None

for i, (ax, (model, psi, T_grid, exp_val, label)) in enumerate(zip(axes, simulations)):
    
    probability_density = np.abs(psi)**2
    
    # Heatmap
    im = ax.imshow(
        probability_density.T,
        aspect="auto", 
        origin="lower",
        extent=[model.ti, model.simulation_time, model.a, model.b], 
        cmap="viridis",
        vmin=0, 
        vmax=global_max_psi2
    )

    # Mean Position <x>
    ax.plot(T_grid, exp_val, color="white", linewidth=2, label=r"$\langle x \rangle(t)$")

    # Equilibrium Position q0(t)
    # Note: q0(t) = t / model.tf
    ax.plot(T_grid, T_grid/model.tf, color="red", linestyle="dashed", linewidth=1.5, label=r"$q_0(t) = t/t_f$")

    # Formatting
    ax.set_ylabel(r"$x$")
    ax.set_ylim(-2, 5)
    
    # Add legend
    ax.text(0.02, 0.9, label, transform=ax.transAxes, color="white", fontweight="bold", bbox=dict(facecolor="black", alpha=0.5, edgecolor="none"))
    if i == 0:
        ax.legend(loc="lower right", frameon=True, facecolor="white", framealpha=0.2)

axes[-1].set_xlabel(r"$t$")

# Place the colorbar
cbar = fig.colorbar(im, ax=axes, orientation="vertical", fraction=0.05, pad=0.02)
cbar.set_label(r"$|\psi(x, t)|^2$")

plt.tight_layout
plt.savefig("heatmap_comaprison.pdf", bbox_inches='tight')
plt.show()
```
