## Ex.1 plotter

### Python script to plot the results obtained from Ex1-Gasparotto-CODE.f90

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Latex-like font for plots
# Note: if LaTeX is not installed, do not run this cell
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "cmr10",
    "text.latex.preamble": r"\usepackage{amsmath, amssymb}",
    "font.size": 16})
```

```python
# Store the correct files depending on how we ran the fortran code:
base_files = [
    Path("row_col.dat"),
    Path("col_row.dat"),
    Path("matmul.dat")
]

optimization_files = [
    Path("results/row_col_O0.dat"),
    Path("results/col_row_O0.dat"),
    Path("results/matmul_O0.dat")
]

# Check file existence
if all(file.exists() for file in base_files):
    data1 = np.loadtxt(base_files[0])
    data2 = np.loadtxt(base_files[1])
    data3 = np.loadtxt(base_files[2])

elif all(file.exists() for file in optimization_files):
    data1 = np.loadtxt(optimization_files[0])
    data2 = np.loadtxt(optimization_files[1])
    data3 = np.loadtxt(optimization_files[2])

else:
    raise TypeError("Error: files does not exists!")
```

```python
# Plot the base case without optimization flags
fig, ax = plt.subplots(figsize=(7, 5))

ax.set_yscale('log')
ax.scatter(data1[:,0], data1[:,1], c="purple", label='Row-by-Column', alpha=0.8, marker='s', edgecolor="black", s=80)
ax.plot(data1[:,0], data1[:,1], c="purple", linestyle='solid', linewidth=3, alpha=0.8)
ax.scatter(data2[:,0], data2[:,1], c="gold", label='Column-by-Row', alpha=0.8, marker='^', edgecolor="black", s=80)
ax.plot(data2[:,0], data2[:,1], c="gold", linestyle='dashed', linewidth=3, alpha=0.8)
ax.scatter(data3[:,0], data3[:,1], c="black", label='Matmul function', alpha=0.8, marker='o', edgecolor="black", s=80)
ax.plot(data3[:,0], data3[:,1], c="black", linestyle='dotted', linewidth=3, alpha=0.8)

ax.set_title('Matrix multiplication time comparison', fontsize=16)
ax.set_xlabel('Matrix Dimension', fontsize=16)
ax.set_ylabel('log(time) (s)', fontsize=16)

ax.legend(frameon=False, loc="best")

fig.tight_layout()
plt.show()
fig.savefig('matrix_multiplication_times_log.pdf', dpi=300, bbox_inches='tight')
```

```python
# Plot the base case without optimization flags

# Same plot but check for the difference in time between row-by-column and column-by-row
# for big dimensions
fig, ax = plt.subplots(figsize=(7, 5))

#ax.set_yscale('log')
ax.scatter(data1[:,0], data1[:,1], c="purple", label='Row-by-Column', alpha=0.8, marker='s', edgecolor="black", s=80)
ax.plot(data1[:,0], data1[:,1], c="purple", linestyle='solid', linewidth=3, alpha=0.8)
ax.scatter(data2[:,0], data2[:,1], c="gold", label='Column-by-Row', alpha=0.8, marker='^', edgecolor="black", s=80)
ax.plot(data2[:,0], data2[:,1], c="gold", linestyle='dashed', linewidth=3, alpha=0.8)
ax.scatter(data3[:,0], data3[:,1], c="black", label='Matmul function', alpha=0.8, marker='o', edgecolor="black", s=80)
ax.plot(data3[:,0], data3[:,1], c="black", linestyle='dotted', linewidth=3, alpha=0.8)

ax.set_xlim(2000, 5000)

ax.set_title('Matrix multiplication time comparison', fontsize=16)
ax.set_xlabel('Matrix Dimension', fontsize=16)
ax.set_ylabel('time (s)', fontsize=16)

ax.legend(frameon=False, loc="best")

fig.tight_layout()
plt.show()
fig.savefig('matrix_multiplication_times_linear.pdf', dpi=300, bbox_inches='tight')
```

```python
# Store the files of the results with the optimization flags, if they exists
all_files = [
    Path("results/row_col_O0.dat"),
    Path("results/col_row_O0.dat"),
    Path("results/matmul_O0.dat"),  

    Path("results/row_col_O2.dat"),
    Path("results/col_row_O2.dat"),
    Path("results/matmul_O2.dat"),

    Path("results/row_col_O3.dat"),
    Path("results/col_row_O3.dat"),
    Path("results/matmul_O3.dat")
]

if all(file.exists() for file in all_files):
    data1_O0 = np.loadtxt(all_files[0])
    data2_O0 = np.loadtxt(all_files[1])
    data3_O0 = np.loadtxt(all_files[2]) 

    data1_O1 = np.loadtxt(all_files[3])
    data2_O1 = np.loadtxt(all_files[4])
    data3_O1 = np.loadtxt(all_files[5])

    data1_O2 = np.loadtxt(all_files[6])
    data2_O2 = np.loadtxt(all_files[7])
    data3_O2 = np.loadtxt(all_files[8])

    # Plot the results
    fig, ax = plt.subplots(figsize=(10, 6))

    colors = {
        'O0': 'black',     
        'O2': 'purple',     
        'O3': 'gold'   
    }

    markers = {
        'Row-by-Column': 's',
        'Column-by-Row': '^',
        'Matmul Function': 'o'
    }

    linestyles = {
        'Row-by-Column': '-',
        'Column-by-Row': '--',
        'Matmul Function': ':'
    }

    # Parameters
    data_dict = {
        'O0': [data1_O0, data2_O0, data3_O0],
        'O2': [data1_O1, data2_O1, data3_O1],
        'O3': [data1_O2, data2_O2, data3_O2],
    }
    methods = ['Row-by-Column', 'Column-by-Row', 'Matmul Function']
    linewidths = [2, 4, 3]
    markersizes = [10, 14, 10]
    alphas = [0.7, 0.8, 1]

    # Plot
    for opt, data_list in data_dict.items():
        for method, data, linewidth, markersize, alpha in zip(methods, data_list, linewidths, markersizes, alphas):
            ax.plot(
                data[:,0], data[:,1],
                linestyle=linestyles[method],
                color=colors[opt],
                marker=markers[method],
                markersize=markersize,
                markeredgecolor="black",
                linewidth=linewidth,
                alpha=alpha,
                label=f'{method} ({opt})'
            )


    ax.set_yscale('log')
    ax.set_xlabel('Matrix dimension', fontsize=16)
    ax.set_ylabel('log(time) (s)', fontsize=16)
    ax.set_title('Optimization level comparison', fontsize=16)

    handles, labels = ax.get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    ax.legend(
        by_label.values(), by_label.keys(),
        fontsize=16,
        loc="best",
        ncol=2,
        frameon=False
    )

    fig.tight_layout()
    plt.show()
    fig.savefig('matrix_multiplication_optimization_comparison.pdf', dpi=300, bbox_inches='tight')
    
else:
    raise TypeError("Error: optimization files does not exist. " \
                    "Compile the fortran file with: ./optimization.sh")
```
